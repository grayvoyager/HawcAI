```python
import torch
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='weights/yolov5ss.pt', force_reload=True)
```

```python
path =('th.jpg')
```

```python
result = model(path)
result.render
```

```python
plt.imshow(np.squeeze(result.render()))
plt.show()
```

```python
#video

cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    x = results.pandas().xyxy[0]
    crops = results.crop
    
    print(x)
    #print(x['class'])
    #print(str(x['name'])== "person")


    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
